## Backround Information

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; Perhaps, the most important topic discussed in Chapters 1-3 is the Sturm-Liouville Theory. It has been described by many as the most useful result ever devised by mathematicians. Many of the summed series that you find in tables of series can be traced back to the Sturm-Lioville Theory. Understanding this application is essential to your understanding of the Equations of Mathematical Physics. The linear partial differential equations discussed later in this course are often referred to as the Equations of Mathematical Physics. The following problem has been designed to step you through the process of application of this theory. 

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;This is a specific application of the theory that allows you to see exactly how the theory leads to a solution. At the same time as you are carrying out the process of solution, think about how many other eigenvalue problems could apply to the same series expansion. You should be able to see that there are infinitely many ways to expand the same function into a series of eigenfunctions that result from the solution of eigenvalue problems. It is a remarkable conclusion that all of these series will converge to the same function on the interval in question. These constructions are formal in the sense that any series expansion is just a mathematical artifact until its convergence has been established. The question of convergence of a series of eigenfunctions has been settled by mathematicians, but not discussed to any degree in the book. That subject has been left for students who want to pursue the subject further. In the mean time, we can make a shallow attempt to verify convergence by summing the series pointwise numerically. This should give us some insight into the rate of convergence and its replication of the function on the interval in question.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;By solving the eigenvalue problem, we mean find the eigenvalues, $\lambda_{n}$, and the nontrivial eigenfunctions, $y_n (x)$, that satisfy the boundary conditions and the differential equation. As you will see when we turn to the Heat Equation, a Parabolic Partial Differential Equation for which the boundary condition at $x=2$, has a very significant physical interpretation within the context of a one-dimensional heat flow problem.

## Instructions:
    
Solve the eigenvalue problem for $y=y(x)$ given below on the interval $0 < x < 2$. <br>

$$ y^{\prime\prime} + 4y^{\prime} + \lambda y \tag{Eq. 1} $$ <br>
$$ y(0) = 0, y(2) + \frac{1}{2} y^{\prime}(2) = 0 \tag{Eq. 2} $$

## Eigenvalue Problem Solution

$$ y^{\prime\prime} + 4y^{\prime} + \lambda y \tag{Eq. 2} $$ <br>

$$ r^{2} + 4r + \lambda = 0 ~~~~\Rightarrow~~~~ r = \frac{-4 ~\pm~ \sqrt{ \left( 4\right )^2 - (4)(1)(\lambda)}}{2(1)} ~~~~\Rightarrow~~~~ r = -2 ~\pm~ \sqrt{4 - \lambda} \tag{Eq. 4} $$ 

Now, we will analyze three seperate scenarios in which the roots exist. 

$$ Assume: 4 - \lambda = \omega ^{2} ~~~~\Rightarrow~~~~ r_{i} = -2 ~\pm~ \omega \tag{Eq. 5} $$ <br>

---

#### Case 1: $~~~\omega^{2} > 0 $ <br>

Assume the solution to be: $ y = e^{r_{i}x} $

$$ y(x) = C_{1}e^{(-2 + \omega)x} + C_{2}e^{(-2 - \omega)x} \tag{Eq. 6} $$ <br>

Take the derivative of equation 6:

$$ y^\prime (x) = C_{1}(-2 + \omega)e^{(-2 + \omega)x} + C_{2}(-2 - \omega)e^{(-2 - \omega)x} \tag{Eq. 7} $$ <br>

Apply the first boundary condition to equation 6:

$$ y (x = 0)=0 \rightarrow C_{1}e^{(-2 + \omega)(0)} + C_{2}e^{(-2 - \omega)(0)} = C_{1} + C_{2} = 0 \tag{Eq. 8} $$ <br>

Apply the second boundary condition to equations 6 and 7:

$$ y(2) + \frac{1}{2} y^{\prime}(2) = 0 \rightarrow C_{1}e^{(-2 + \omega)(2)} + C_{2}e^{(-2 - \omega)(2)} + \frac{1}{2} \left [ C_{1}(-2 + \omega)e^{(-2 + \omega)(2)} + C_{2}(-2 - \omega)e^{(-2 - \omega)(2)} \right ] = 2 C_{2} e^{4-2\omega} = 0 \tag{Eq. 9} $$

$$ \therefore C_{2} = 0 ~and~ C_{1} = 0 $$ <br>

From these results, we can see only trivial solutions exist for this case.

---

#### Case 2: $~~~\omega^{2} = 0 $ <br>

Again, assume the solution to be: $ y = e^{r_{i}x} $

$$ y(x) = C_{1}e^{-2x} + C_{2}xe^{-2x} \tag{Eq. 10} $$ <br>

Take the derivative of equation 10:

$$ y(x) = -2 C_{1}e^{-2x} + C_{2}e^{x} + 2C_{2}xe^{2x} \tag{Eq. 11} $$ <br>

Apply the first boundary condition to equation 10:

$$ y(x = 0) = 0 \rightarrow y(0) = C_{1}e^{-2(0)} + C_{2}(0)e^{-2(0)} = 0 ~~~~\Rightarrow~~~~ C_{1} = 0 \tag{Eq. 12} $$ <br>

Apply the second boundary condition to equations 10 and 11:

$$ y(x = 2) + \frac{1}{2} y^{\prime}(x = 2) = 0 \rightarrow 2 C_{2}e^{2(2)} + \frac{1}{2} \left [ C_{2}e^{2} + 2C_{2}(2)e^{2(2)} \right ] = 2C_{2}e^{2} \left [ e^{2} + \frac{1}{4} + e^{2} \right ] = 0 \tag{Eq. 13} $$

$$ \therefore C_{2} = 0 $$

Again, from these results, we can see only trivial solutions exist for this case.

---

#### Case 3:  $~~~-\omega^{2} < 0 $ <br>

It is important to note that since $\omega^{2}$ is negative, an imaginary solution exists. <br>

$$ y(x) = C_{1}e^{-2x}cos(\omega x) + C_{2}e^{-2x}sin(\omega x) \tag{Eq. 14} $$ <br>

This time, we will use the first boundary condition on equation 14 to make the problem more simple.

$$ y(x = 0) = 0 \rightarrow y(0) = C_{1}e^{-2(0)} cos \left[ \omega (0) \right] + C_{2}e^{-2(0)} sin \left[ \omega (0) \right] = 0 ~~~~\Rightarrow~~~~ C_{1} = 0 \tag{Eq. 15} $$ <br>

Take the derivative of equation 14:

$$ y(x) = -2C_{2}e^{-2x}sin(\omega x) + C_{2} \omega e^{-2x} cos (\omega x) \tag{Eq. 16} $$ <br>

Apply the second boundary condition to equations 14 and 16:

$$ y(x = 2) + \frac{1}{2} y^{\prime}(x = 2) = 0 \rightarrow C_{2}e^{-2(2)}sin(2\omega) + \frac{1}{2} \left[ -2C_{2}e^{-2(2)}sin(2\omega) + C_{2}\omega e^{-2(2)}cos(2\omega)\right ] = 0 $$ 

$$~~~~\Rightarrow~~~~ C_{2}\omega e^{-4}cos(2\omega) = 0 \tag{Eq. 17} $$ 

Here, $\omega , e^{-4}, ~nor~ C_{2}$ can be equal to zero, or it will drive to another trivial solution... <br><br>


The only logical determination is that $ cos(2\omega) = 0$.

$$ cos(2\omega) = 0 ~~~at~~~ 2\omega = \frac{2n-1}{2}\pi ~~~~\Rightarrow~~~~ \omega = \frac{2n-1}{4} \pi \tag{Eq. 18}$$ <br>



---

#### Write the Eigenvalues and Eigenfunctions

<br> <center> $ y_{n}(x) = e^{-2x}sin \left ( \frac{2n-1}{4}\pi x \right ) \tag{Eq. 19} $ ...from equation 14 (Note: the constant multiplyer can be left off.<br><br>

<center> $ \lambda_{n} = \left [ \frac{2n-1}{4}\pi \right ]^2 +4 \tag{Eq. 20} $ ...from equation 5. 

## Expanding the Eigenfunction

Now, we need to expand the eigenfunction (equation 19) on the scale from the background...

To expand the equation, we will use the following relation:

$$ f(x) = \sum_{n=1}^{\infty} C_{m}y_{m}(x) $$

To solve for the constants, $C_{n}$, we will use the following relation:
    
$$ C_{m} = \frac{\int_{0}^{2} f(x) w(x) y_m(x) ~dx}{\int_{0}^{2} w(x) y_{m}^{2} ~dx}, ~~~w(x) = e^{4x} ~~~~\Rightarrow~~~~ C_{m} = \frac{\int_{0}^{2} e^{4x} e^{-2x}sin \left ( \frac{2m-1}{4}\pi x \right ) ~dx}{\int_{0}^{2} e^{4x} \left [ e^{-2x}sin \left ( \frac{2m-1}{4}\pi x \right ) \right ]^2 ~dx} ~~~~\Rightarrow~~~~ C_{m} = \int_{0}^{2} e^{2x}sin \left ( \frac{2m-1}{4}\pi x \right )~dx $$

---

## Demonstrating the Expansion Converges to a Value of 1

Assume: $ \beta = \frac{2m-1}{4} \pi $

Our matrix (for storing the values) will be in the following order:

<center>

| x value        | m value      | $\beta$ value | $C_{m}*y_{m}(x_{1})$ | $C_{m}*y_{m}(x_{2})$ | ...          |
|:-----------:   |:-----------: |:-----------:  |:-----------:         |:-----------:         |:-----------: |
|$x_{init}$      | 1            | $\beta_{1}$   |    $solution_{1}$    |  $solution_{1}$      | ...          |
|$x_{init}$+step | 2            | $\beta_{2}$   |    $solution_{2}$    |  $solution_{1}$      | ...          |
    
</center>

for values of m from 1 to 0. <br><br><br>

#### Define our Functions

```python
import math

#---------------------------------------------------------------------------%

def beta_func(mval):
    
    beta = ((2*mval)-1)*(math.pi/4)
    
    return(beta)

#---------------------------------------------------------------------------%

def y_func(betaval, xval):
    
    y = math.exp(-2*xval)*math.sin(betaval*xval)
    
    return(y)

#---------------------------------------------------------------------------%

def C_func (betaval):
    
    C = ((-math.exp(4)*betaval*math.cos(2*betaval))+(2*math.exp(4)*math.sin(2*betaval))+(betaval))/((betaval**2)+4)
    
    return(C)

#---------------------------------------------------------------------------%

def f_func (yval, Cval):
    
    f = Cval*yval
    
    return(f)
```

```python
import numpy as np

MATRIX = np.zeros((1050,6), dtype = float)
```

```python
MATRIX_step_pos = 0

x = 0

while x < 2.1:

    for m in range (1,51):

        BETA = beta_func(m)
        CM = C_func(BETA)
        YM = y_func(BETA, x)
        F = f_func(YM, CM)

        MATRIX[MATRIX_step_pos,0] = x
        MATRIX[MATRIX_step_pos,1] = m
        MATRIX[MATRIX_step_pos,2] = BETA
        MATRIX[MATRIX_step_pos,3] = CM
        MATRIX[MATRIX_step_pos,4] = YM
        MATRIX[MATRIX_step_pos,5] = F
        
        MATRIX_step_pos += 1
        
    x += 0.1
```
